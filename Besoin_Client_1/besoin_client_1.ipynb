# Besoin Client 1 – Visualisation sur carte

Objectif : visualiser les bornes IRVE sur une carte interactive par type d'implantation et créer une heatmap de densité spatiale.

Responsable : **Personne 1**

## 1. Imports et chargement des données

```python
import pandas as pd
import numpy as np
import plotly.express as px
import folium
from folium.plugins import HeatMap, MarkerCluster
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('IRVE_cleaned.csv', low_memory=False)
print(df.shape)
df.head()
```

## 2. Préparation des données

**Justification des colonnes choisies :** (à remplir)

On sélectionne les colonnes pertinentes pour la visualisation géographique.

```python
# Colonnes utiles pour BC1 – à adapter selon votre export Big Data
cols = ['consolidated_latitude', 'consolidated_longitude',
        'implantation_station', 'puissance_nominale']

df_map = df[cols].dropna()

# Vérification
print(f'Lignes après nettoyage : {len(df_map)}')
print(df_map.dtypes)
df_map['implantation_station'].value_counts()
```

## 3. Distribution des types d'implantation

```python
fig = px.bar(
    df_map['implantation_station'].value_counts().reset_index(),
    x='implantation_station', y='count',
    title="Distribution des types d'implantation",
    labels={'implantation_station': "Type d'implantation", 'count': 'Nombre de bornes'}
)
fig.show()   
```

## 4. Carte interactive par type d'implantation

**Justification du choix de la bibliothèque :** (à remplir – ex. Folium car intégration Leaflet.js, export HTML)

```python
# Palette de couleurs par type d'implantation
types = df_map['implantation_station'].unique()
palette = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'cadetblue']
color_map = {t: palette[i % len(palette)] for i, t in enumerate(types)}

carte = folium.Map(
    location=[df_map['consolidated_latitude'].mean(),
               df_map['consolidated_longitude'].mean()],
    zoom_start=6
)

cluster = MarkerCluster().add_to(carte)

for _, row in df_map.sample(min(3000, len(df_map))).iterrows():
    folium.CircleMarker(
        location=[row['consolidated_latitude'], row['consolidated_longitude']],
        radius=4,
        color=color_map.get(row['implantation_station'], 'gray'),
        fill=True,
        popup=row['implantation_station']
    ).add_to(cluster)

carte.save('carte_implantation.html')
print('Carte sauvegardée : carte_implantation.html')
carte
```

## 5. Heatmap de densité spatiale

```python
heatmap_carte = folium.Map(
    location=[df_map['consolidated_latitude'].mean(),
               df_map['consolidated_longitude'].mean()],
    zoom_start=6
)

heat_data = df_map[['consolidated_latitude', 'consolidated_longitude']].values.tolist()
HeatMap(heat_data, radius=10, blur=15, max_zoom=1).add_to(heatmap_carte)

heatmap_carte.save('heatmap_densite.html')
print('Heatmap sauvegardée : heatmap_densite.html')
heatmap_carte
```
